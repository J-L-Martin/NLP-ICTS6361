# Files 
two ways to handle files:

1. read line by line 
2. read the whole file 

```python
# read line by line
fname = '../corpus/bbc.com_20190409_titles_small.txt'
fhandle = open(fname, encoding='utf-8')
for line in fhandle:
    print(line.strip())
```

```python
text = open(fname, encoding='utf-8').read()
```
