# Python Files

# read file line by line 

```python
fname = '../corups/bbc.com_20190409_titles_small.txt'
fhandle = open(fname, encoding='utf-8')
for line in fhandle:
    print(line.strip())
```

# read the whole file 

```python
text = open(fname, encoding='utf-8').read()
```
